  **1. (For Data cleaning)**

  **Find all null values in the dataset. If there is any null value in any column, then fill with the mean of that column.**

```python
df.Cylinders.fillna(df.Cylinders.mean(), inplace = True)
```

 **2. (Value Counts)**
  
  **Check what are the difference types of 'Make' are there in our dataset.What is the count(occurrence) of each 'Make' in the data?**

```python
df.groupby(df.Make).sum()
```

  **3. (Filter)**
  **Show all the recordes where 'Origin' is Asia or europe.**


```python
df[df.Origin.isin(['Asia', 'europe'])]
```

  **4. (Removing unwanted records)**
  **Remove all the records where 'Weight' is above 4000**

```python
df = df[~(df.Weight > 4000)]
df
```

 **5. (Applying function on  a column, add some value)**
  **Incease all the values of 'MPG_City' column by 3**

```python
df.MPG_City += 3
```
